## **Paradigma Conexionista - Rede Neural**

```python
df4 = pd.read_pickle('df_social_data_train.pkl')
df4 = df4.dropna()
df4
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
df4_train, df4_test = train_test_split(df4, test_size=0.3, random_state=42, stratify=df4.engagement)
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X4_train = vectorizer.fit_transform(df4_train['content']).toarray()
X4_test = vectorizer.transform(df4_test['content']).toarray()
```

```python
#Encode dos rótulos (low/high) para 0/1
le = LabelEncoder()
y4_train = le.fit_transform(df4_train['engagement'])
y4_test = le.transform(df4_test['engagement'])
```

```python
#Construção do modelo MLP
model = Sequential([
    Dense(128, activation='relu', input_shape=(X4_train.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X4_train, y4_train, epochs=10, batch_size=32, validation_split=0.1)
```

```python
y4_pred_prob = model.predict(X4_test)
y4_pred = (y4_pred_prob > 0.5).astype(int)

print(classification_report(y4_test, y4_pred, target_names=le.classes_))
```

Por outro lado, o paradigma conexionista (Rede Neural) apresenta uma acurácia um pouco menor (79%), porém um melhor equilíbrio no recall entre as classes, conseguindo identificar uma maior proporção dos exemplos da classe “high”, embora com precisão ainda limitada. Esses resultados indicam que, apesar das redes neurais conseguirem lidar melhor com o desbalanceamento e capturar nuances da classe minoritária, modelos estatísticos e probabilísticos tendem a favorecer a classe majoritária, o que pode ser um problema em cenários onde a detecção da classe minoritária é crítica.

# **Atividade 2**

Vamos utilizar o K-Means para encontrar padrões, com análise e visualização.

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Gerar embeddings
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(df['content'].tolist(), show_progress_bar=True)
```

```python
# 2. Encontrar o número ideal de clusters usando silhouette score
range_n_clusters = list(range(2, 11))
silhouette_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(embeddings)
    score = silhouette_score(embeddings, cluster_labels)
    silhouette_scores.append(score)
    print(f"Silhouette score para {n_clusters} clusters: {score:.4f}")
```

```python
# Plotar silhouette scores
plt.figure(figsize=(8,5))
sns.lineplot(x=range_n_clusters, y=silhouette_scores, marker='o')
plt.title('Silhouette Score para diferentes números de clusters')
plt.xlabel('Número de clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
# 3. Escolher o melhor número de clusters (exemplo: o que tem maior silhouette)
best_k = range_n_clusters[np.argmax(silhouette_scores)]
print(f"Número ideal de clusters: {best_k}")
```

```python
# 4. Aplicar k-means com o melhor número de clusters
kmeans = KMeans(n_clusters=best_k, random_state=42)
df['cluster'] = kmeans.fit_predict(embeddings)
```

```python
# 5. Analisar distribuição dos clusters
print("\nDistribuição de posts por cluster:")
print(df['cluster'].value_counts())
```

```python
# 6. Ver relação cluster x engajamento
print("\nContagem de engajamento por cluster:")
print(pd.crosstab(df['cluster'], df['engagement']))
```

```python
# 7. Palavras-chave por cluster
print("\nPalavras-chave mais frequentes por cluster:")

for cluster_id in range(best_k):
    texts = df[df['cluster'] == cluster_id]['content']
    vectorizer = TfidfVectorizer(stop_words='english', max_features=10)
    X = vectorizer.fit_transform(texts)
    keywords = vectorizer.get_feature_names_out()
    print(f"Cluster {cluster_id}: {', '.join(keywords)}")
```

```python
# 8. Visualização dos clusters com PCA + t-SNE

# Redução dimensional com PCA para acelerar t-SNE
pca = PCA(n_components=50, random_state=42)
embeddings_pca = pca.fit_transform(embeddings)

# t-SNE para 2D
tsne = TSNE(n_components=2, random_state=42, perplexity=40, n_iter=300)
embeddings_tsne = tsne.fit_transform(embeddings_pca)

# Criar DataFrame para plot
df_vis = pd.DataFrame()
df_vis['tsne-2d-one'] = embeddings_tsne[:,0]
df_vis['tsne-2d-two'] = embeddings_tsne[:,1]
df_vis['cluster'] = df['cluster'].astype(str)
df_vis['engagement'] = df['engagement']

plt.figure(figsize=(10,7))
sns.scatterplot(x='tsne-2d-one', y='tsne-2d-two',
                hue='cluster',
                style='engagement',
                palette='tab10',
                data=df_vis,
                alpha=0.7)
plt.title('Visualização dos clusters com t-SNE (colorido por cluster, shape por engajamento)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

# **Atividade 3**

Agente Reescritor com Aprendizado Supervisionado
Entrada: uma postagem qualquer com low engagement.

Saída: uma versão reescrita no estilo de postagens com high engagement.

Como fazer:

Use embeddings para encontrar os k posts mais similares com high engagement.

Use uma LLM (como GPT ou local com Llama) para reescrever o post no estilo dos mais engajadores.

Ex: "Aqui está uma versão mais engajadora baseada em outros posts parecidos que tiveram muito sucesso."

**Estrutura do Projeto**
3 etapas:

**Preparação (Embeddings)**: Converter todos os posts de alto engajamento em vetores numéricos (embeddings) para que possamos calcular similaridade. Isso é feito apenas uma vez.

**Busca (Retrieval)**: Criar uma função que, dado um post de baixo engajamento, encontra os k posts mais parecidos no nosso conjunto de vetores de alto engajamento.

**Geração (Generation)**: Criar a função principal do agente, que usa os posts encontrados como inspiração para a LLM reescrever o post original.

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np


# 1. Separar os posts por engajamento
high_engagement_df = df[df['engagement'] == 'high'].copy()
low_engagement_df = df[df['engagement'] == 'low'].copy()

print(f"Temos {len(high_engagement_df)} posts com alto engajamento.")
print(f"Temos {len(low_engagement_df)} posts com baixo engajamento.")

# 2. Carregar o modelo de embeddings
print("Carregando modelo de embeddings...")
model = SentenceTransformer('all-MiniLM-L6-v2')

# 3. Gerar os embeddings para os posts de ALTO engajamento
print("Gerando embeddings para os posts de alto engajamento...")
high_engagement_embeddings = model.encode(high_engagement_df['content'].tolist(), show_progress_bar=True)

# 4. Criar um índice FAISS para busca rápida
# O índice FAISS vai armazenar os vetores e nos permitir buscar por similaridade de forma eficiente.
print("Criando índice FAISS...")
embedding_dim = high_engagement_embeddings.shape[1]
index = faiss.IndexFlatL2(embedding_dim) # Usando a distância L2 (similar à Euclidiana)
index.add(high_engagement_embeddings)

print("\n✅ Preparação concluída! O índice com os posts de alto engajamento está pronto.")
```

```python
def encontrar_posts_similares(texto_post_original, k=3):
    """
    Encontra os k posts de alto engajamento mais similares a um texto de entrada.
    """
    # Gera o embedding para o post de entrada
    post_embedding = model.encode([texto_post_original])

    # Busca no índice FAISS pelos k mais próximos
    # D: distâncias, I: índices dos posts no high_engagement_df
    distances, indices = index.search(post_embedding, k)

    # Retorna o conteúdo dos posts encontrados
    similar_posts = high_engagement_df['content'].iloc[indices[0]].tolist()
    return similar_posts
```

```python
from huggingface_hub import InferenceClient

# Use o cliente que já configuramos e testamos
HF_TOKEN = "SEU_TOKEN_COM_PERMISSAO_DE_WRITE"
client = InferenceClient(model="mistralai/Mistral-7B-Instruct-v0.3", token=HF_TOKEN)

def agente_reescritor(post_original, k=3):
    """
    Agente que reescreve um post de baixo engajamento usando posts de
    alto engajamento como inspiração.
    """
    print(f"📝 Post Original:\n'{post_original}'")

    # 1. Busca: Encontrar posts similares de alto engajamento
    print("\n🔍 Buscando posts de sucesso similares...")
    exemplos_sucesso = encontrar_posts_similares(post_original, k)

    print(f"✨ Encontrados {len(exemplos_sucesso)} exemplos para inspiração.")

    # 2. Geração: Construir o prompt e chamar a LLM
    print("\n🧠 Acionando a LLM para reescrever o post...")

    # Formatando os exemplos para incluir no prompt
    exemplos_formatados = "\n".join([f"- '{ex}'" for ex in exemplos_sucesso])

    prompt = f"""
Você é um especialista em marketing de redes sociais e um copywriter criativo. Sua tarefa é reescrever um post de baixo engajamento para torná-lo mais atrativo e com maior potencial de viralizar.

**Post Original (baixo engajamento):**
"{post_original}"

**Exemplos de posts com tema similar que tiveram alto engajamento:**
{exemplos_formatados}

**Sua Tarefa:**
Com base nos exemplos de sucesso, reescreva o post original. Mantenha a ideia central, mas use um tom, estilo e estrutura que incentive mais reações e comentários. Seja criativo, direto e impactante.

**Resposta (apenas o novo post):**
"""

    response = client.chat_completion(
        messages=[{"role": "user", "content": prompt}],
        max_tokens=250, # Aumentei um pouco para dar mais liberdade criativa
        temperature=0.75,
        top_p=0.9,
    )

    post_reescrito = response.choices[0].message.content.strip()
    return post_reescrito
```

```python
# Pegando um post aleatório de baixo engajamento para testar
post_para_melhorar = low_engagement_df.sample(1)['content'].iloc[0]

# Chamar o nosso agente criativo
novo_post = agente_reescritor(post_para_melhorar, k=3)

print("\n\n🚀 Post Reescrito com Potencial de Engajamento!\n")
print("="*50)
print(novo_post)
print("="*50)
```
